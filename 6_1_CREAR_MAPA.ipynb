```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("dataset_general.csv", encoding = "latin-1")
dataset_2019 = pd.read_csv("dataset_2019.csv", encoding = "latin-1")
dataset_2020 = pd.read_csv("dataset_2020.csv", encoding = "latin-1")
dataset_2021 = pd.read_csv("dataset_2021.csv", encoding = "latin-1")
dataset_2022 = pd.read_csv("dataset_2022.csv", encoding = "latin-1")
```

```python
# Función para agregar la fila de México (País)
def agregar_fila_pais(df, columna_estado='estado'):
    # Separar las columnas numéricas para evitar sumar columnas no numéricas
    columnas_numericas = df.select_dtypes(include=['number']).columns
    # Crear una nueva fila con los valores sumados de las columnas numéricas
    fila_mexico = df[columnas_numericas].sum(axis=0).round(2)
    # Convertir la fila a un diccionario para agregarla
    nueva_fila = fila_mexico.to_dict()
    # Añadir la nueva fila al DataFrame
    df.loc[len(df)] = nueva_fila
    # Cambiar el valor de la columna 'Estado' en la última fila
    df.loc[df.index[-1], columna_estado] = "México_País"

# Función para calcular proporciones redondeadas a 2 decimales
def calcular_proporciones(df):
    df['proporcion_droga'] = ((df['droga_1'] / df['total_observaciones']) * 100).round(3)
    df['sentimiento_negativo_droga'] = ((df['sentimiento_0'] / df['droga_1']) * 100).round(2)
    df['sentimiento_neutral_droga'] = ((df['sentimiento_1'] / df['droga_1']) * 100).round(2)
    df['sentimiento_positivo_droga'] = ((df['sentimiento_2'] / df['droga_1']) * 100).round(2)

# Aplicar la función de agregar fila de México (País) y luego calcular proporciones
agregar_fila_pais(dataset)
agregar_fila_pais(dataset_2019)
agregar_fila_pais(dataset_2020)
agregar_fila_pais(dataset_2021)
agregar_fila_pais(dataset_2022)

calcular_proporciones(dataset)
calcular_proporciones(dataset_2019)
calcular_proporciones(dataset_2020)
calcular_proporciones(dataset_2021)
calcular_proporciones(dataset_2022)
```

```python
import geopandas as gpd
import pandas as pd

# Ruta al archivo shapefile
ruta_shp = r'C:\Users\ANDRES.PEREZ\Documents\PERSONAL\maestria\TESIS\Tesis\SCRIPTS\mapas\mexican-states.shp'
shapefile = gpd.read_file(ruta_shp)

# Diccionario de correspondencias
correspondencias = {
    "Estado de México": "México",
    "Coahuila": "Coahuila de Zaragoza",
    "Veracruz": "Veracruz de Ignacio de la Llave",
    "Michoacán": "Michoacán de Ocampo",
    "Ciudad de México": "Ciudad de México",
    "Baja California": "Baja California",
    "Baja California Sur": "Baja California Sur",
    "San Luis Potosí": "San Luis Potosí",
    "Querétaro": "Querétaro",
    "Quintana Roo": "Quintana Roo",
    # Agregar otras correspondencias si es necesario
}

# Lista de datasets
datasets = {
    "2019": dataset_2019,
    "2020": dataset_2020,
    "2021": dataset_2021,
    "2022": dataset_2022,
}

# Normalizar valores y verificar coincidencias para cada dataset
for year, current_dataset in datasets.items():
    # Crear una nueva columna con valores normalizados
    current_dataset["estado_normalizado"] = current_dataset["estado"].replace(correspondencias)
    
    # Verificar coincidencias con el shapefile
    coincidencias = shapefile["name"].isin(current_dataset["estado_normalizado"]).sum()
    print(f"Año {year}: Número de estados coincidentes: {coincidencias} de {len(shapefile)}")
```

```python
# Hacer la unión entre shapefile y dataset
shapefile_general = shapefile.merge(
    right=dataset,
    left_on="name",
    right_on="estado_normalizado",
    how="left"
)
# Hacer la unión entre shapefile y dataset_2019
shapefile_2019 = shapefile.merge(
    right=dataset_2019,
    left_on="name",
    right_on="estado_normalizado",
    how="left"
)
# Hacer la unión entre shapefile y dataset_2020
shapefile_2020 = shapefile.merge(
    right=dataset_2020,
    left_on="name",
    right_on="estado_normalizado",
    how="left"
)
# Hacer la unión entre shapefile y dataset_2021
shapefile_2021 = shapefile.merge(
    right=dataset_2021,
    left_on="name",
    right_on="estado_normalizado",
    how="left"
)
# Hacer la unión entre shapefile y dataset_2022
shapefile_2022 = shapefile.merge(
    right=dataset_2022,
    left_on="name",
    right_on="estado_normalizado",
    how="left"
)
```

```python
import matplotlib.pyplot as plt

# Calcular los máximos y mínimos globales para cada columna
def calcular_rango_global(shapefiles, columnas):
    """
    Calcula los valores mínimos y máximos globales para cada columna en los shapefiles.
    
    Args:
        shapefiles (list): Lista de GeoDataFrames.
        columnas (list): Lista de nombres de columnas a evaluar.
    
    Returns:
        dict: Diccionario con el rango global (min, max) para cada columna.
    """
    rangos = {}
    for columna in columnas:
        valores = [data[columna].dropna() for data in shapefiles]
        global_min = min(valores, key=min).min()
        global_max = max(valores, key=max).max()
        rangos[columna] = (global_min, global_max)
    return rangos

# Generar los rangos globales
shapefiles = [shapefile_general]
columnas = ['proporcion_droga', "droga_1",'sentimiento_negativo_droga', 'sentimiento_neutral_droga',
            'sentimiento_positivo_droga', 'total_observaciones']
rangos_globales = calcular_rango_global(shapefiles, columnas)

# Modificar la función para usar límites globales
def crear_mapa(data, columna, titulo, archivo_salida, cmap='Greens', rango=None):
    """
    Genera un mapa temático basado en una columna específica del GeoDataFrame y lo guarda como JPEG.
    
    Args:
        data (GeoDataFrame): Datos geográficos.
        columna (str): Nombre de la columna a graficar.
        titulo (str): Título del mapa.
        archivo_salida (str): Nombre del archivo JPEG de salida.
        cmap (str): Paleta de colores (default: 'Greens').
        rango (tuple): Rango global (min, max) para la escala de colores.
    """
    ax = data.plot(
        column=columna,
        legend=True,
        legend_kwds={
            'label': titulo,
            'orientation': 'horizontal'
        },
        cmap=cmap,
        figsize=(10, 8),
        vmin=rango[0] if rango else None,
        vmax=rango[1] if rango else None
    )
    ax.set_title(titulo, fontsize=16)
    ax.axis('off')  # Opcional, elimina los ejes
    plt.savefig(archivo_salida, format='jpeg', dpi=300, bbox_inches='tight')
    plt.close()  # Cierra la figura para liberar memoria

# Crear y exportar los mapas utilizando los límites globales
for columna in columnas:
    rango = rangos_globales[columna]
    crear_mapa(shapefile_general, columna, f'{columna.replace("_", " ").title()}', f'{columna}.jpeg', cmap='Purples', rango=rango)

# Crear mapas para cada año
for year, shapefile_data in zip(
    ['2019', '2020', '2021', '2022'], 
    [shapefile_2019, shapefile_2020, shapefile_2021, shapefile_2022]
):
    for columna in columnas:
        rango = rangos_globales[columna]
        crear_mapa(
            shapefile_data,
            columna,
            f'{columna.replace("_", " ").title()} ({year})',
            f'{columna}_{year}.jpeg',
            cmap='Purples',
            rango=rango
        )
```
